```python
#!/usr/bin/env python
"""
__author__ = "Param Popat"
__version__ = "4"
"""


import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Embedding


def normalize(a):
    """
    Calculates Z score
    :param a: Array to be normalized
    :return: Normalized Array
    """
    a = (a - np.mean(a)) / (np.std(a))
    return a


def inverse_normalize(a, st_deviation, the_mean):
    """
    Translates normalized values to original values
    :param a: Normalized array
    :param st_deviation: Standard deviation of original levels
    :param the_mean: Mean of original levels
    :return: Original Values
    """
    a = (a * st_deviation) + the_mean
    return a


def get_model(xtrain, ytrain, xtest, ytest):
    """
    Trains and returns the model
    :return: Trained Model
    """
    regressor = Sequential()
    regressor.add(LSTM(units=100, return_sequences=True, input_shape=(xtrain.shape[1], xtrain.shape[2])))
    regressor.add(Dropout(0.2))
    regressor.add(LSTM(units=80, return_sequences=True, activation='relu'))
    regressor.add(Dropout(0.2))
    regressor.add(LSTM(units=40))
    regressor.add(Dense(1, activation='linear'))
    regressor.compile(optimizer='adam', loss='mse')
    regressor.fit(xtrain, ytrain, epochs=10, batch_size=15, validation_data=(xtest, ytest), verbose=2,
                  shuffle=False)
    return regressor


def train(stock, to_plot=False):
    """
    Loads dataset and trains LSTM based model on it.
    :param stock: Name of Dataset
    :param to_plot: True if plots to be shown
    :return: Trained Model
    """

    # Load Dataset
    dataset = pd.read_csv(stock + '.csv').iloc[::-1]
    X = dataset.iloc[:, :-1].values
    y = dataset.iloc[:, -1].values

    std = np.std(dataset)
    mean = np.mean(dataset)

    # Normalize
    temp = normalize(dataset)
    tempX = temp.iloc[:, :-1].values
    tempy = temp.iloc[:, -1].values

    X_train, X_test, y_train, y_test = train_test_split(tempX, tempy, test_size=0.25, shuffle=False)

    # Reshaping
    tempX = np.reshape(tempX, (tempX.shape[0], 1, tempX.shape[1]))
    X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
    X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

    # Initialising the RNN
    regressor = get_model(X_train, y_train, X_test, y_test)

    # Predicting
    y_test_all = regressor.predict(tempX)

    # Translating back to original values
    y_test_all = inverse_normalize(y_test_all, std[std.__len__() - 1], mean[mean.__len__() - 1])

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=False)
    X_train_, X_test_, y_train_pred, y_test_pred = train_test_split(X, y_test_all, test_size=0.25, shuffle=False)

    if to_plot:
        plt.plot(y_train, color='green', label='original')
        plt.plot(y_train_pred, color='navy', label='predicted')
        plt.title('Train')
        plt.legend()
        plt.show()

        plt.plot(y_test, color='green', label='original')
        plt.plot(y_test_pred, color='navy', label='predicted')
        plt.title('Test')
        plt.legend()
        plt.show()

    print('MSE TRAIN = ' + str(mse(y_train, y_train_pred)))
    print('MSE TEST = ' + str(mse(y_test, y_test_pred)))
    print(mae(y_test, y_test_pred))
    return regressor


trained_model = train('NIFTY', to_plot=True)
```
